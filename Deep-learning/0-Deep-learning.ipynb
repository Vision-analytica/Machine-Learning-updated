## Definition
Deep learning is a subfield of machine learning that involves the use of artificial neural networks with multiple layers to model and solve complex problems. 

Artificial Neural Networks (ANNs) are a type of deep learning model that is designed to simulate the way the human brain works. ANNs consist of interconnected nodes (neurons) that transmit signals and perform computations on input data to produce output values. These networks can be used for a wide range of applications, including image recognition, speech recognition, natural language processing, and more.

## Classification

ANNs can be classified into several categories based on their structure and function, including:

1. **Feedforward neural networks:** These are neural networks that have a series of interconnected layers, where the output of each layer serves as the input for the next layer.

2. **Convolutional neural networks (CNNs):** CNNs are primarily used for image and video processing tasks. They consist of layers of convolutional filters that can identify patterns in the input images.

3. **Recurrent neural networks (RNNs):** These are neural networks that are well-suited to sequence analysis tasks, such as natural language processing or speech recognition. They use a type of neural network layer called a recurrent layer that can maintain an internal state and process inputs one at a time.

4. **Generative adversarial networks (GANs):** GANs are a type of network that can generate new data based on a set of input data. They consist of two networks: a generator network that creates new data, and a discriminator network that evaluates the quality of the generated data.

5. **Autoencoders:** Autoencoders are designed to learn a compressed representation of input data. They consist of an encoder network that compresses the input data into a low-dimensional representation, and a decoder network that can reconstruct the original input data from the compressed representation.

## Deep Learning framework

In the context of deep learning, a framework is a software library or tool that provides a set of APIs (application programming interfaces) and abstractions to simplify the development of deep neural networks. Frameworks typically include pre-implemented building blocks for common neural network layers, such as convolutional, recurrent, and fully connected layers, as well as optimization algorithms and training routines.

There are many popular frameworks for deep learning, some of which include:

- **TensorFlow:** Developed by Google, it is an open-source software library for dataflow and differentiable programming across a range of tasks.

- **PyTorch:** Developed by Facebook, it is an open-source machine learning framework used for applications such as computer vision and natural language processing.

- **Keras:** An open-source neural network library written in Python, it runs on top of other deep learning frameworks such as TensorFlow and Theano.

- **Caffe:** Developed by Berkeley Vision and Learning Center, it is a deep learning framework that specializes in image recognition.

- **Theano:** Another popular open-source numerical computation library, it is used for deep learning and other mathematical computations.

- **MXNet:** An open-source deep learning framework that is highly scalable and supports a range of programming languages including Python, R, and Julia.

- **Chainer:** A Python-based, open-source deep learning framework that is highly flexible and allows for dynamic computation graphs.

These frameworks provide a range of features and tools for developing and training deep neural networks, making it easier for developers and researchers to experiment with different architectures and optimize their models for specific tasks.

    - Machine translation systems, like Google Translate, employ deep learning to improve the accuracy and fluency of translations between different languages.

    - Sentiment analysis algorithms analyze text data from social media, customer reviews, or surveys, using deep learning models to determine the sentiment expressed in the text (e.g., positive, negative, neutral).

3. **Medical Diagnostics:** Deep learning has shown promising results in medical imaging analysis and disease diagnostics. Examples include:

    - Deep learning models can detect anomalies and classify medical images, such as X-rays, MRIs, or CT scans, assisting radiologists in diagnosing diseases like cancer or identifying abnormalities.

    - Deep learning algorithms have been used to predict the risk of certain diseases, such as diabetic retinopathy or cardiovascular diseases, based on patient data, enabling early detection and intervention.

    - Streaming platforms like Netflix and Spotify employ deep learning algorithms to recommend personalized movies, TV shows, or music based on a user's viewing or listening history.

    - E-commerce platforms, such as Amazon, utilize deep learning-based recommendation systems to suggest products based on a user's browsing history, purchase behavior, and similar user profiles.

5. **Speech Recognition:** Deep learning has significantly enhanced speech recognition accuracy and enabled voice-controlled applications. Examples include:

    - Voice assistants like Apple's Siri, Amazon's Alexa, or Google Assistant utilize deep learning models to accurately recognize and respond to spoken commands and queries.

    - Transcription services employ deep learning algorithms to convert spoken language into written text, facilitating tasks such as transcription services, voice search, or closed captioning.

```python
import numpy as np
import matplotlib.pyplot as plt

# Quadratic cost function
def cost_function(x):
    return x**2

# Gradient of the cost function
def gradient(x):
    return 2*x

# Gradient Descent
def gradient_descent(learning_rate, iterations):
    x = 10 # Initial guess
    history = [x]
    for i in range(iterations):
        x -= learning_rate * gradient(x)
        history.append(x)
    return history

# Stochastic Gradient Descent
def stochastic_gradient_descent(learning_rate, iterations):
    x = 10 # Initial guess
    history = [x]
    for i in range(iterations):
        gradient_x = gradient(x)
        x -= learning_rate * gradient_x
        history.append(x)
    return history

# Mini-batch Gradient Descent
def mini_batch_gradient_descent(batch_size, learning_rate, iterations):
    x = 10 # Initial guess
    history = [x]
    for i in range(iterations):
        if batch_size > len(history):
            batch_size = len(history)  # Adjust batch size if it's larger than the history length
        random_indices = np.random.choice(len(history), batch_size, replace=False)
        x -= learning_rate * np.mean([gradient(history[j]) for j in random_indices])
        history.append(x)
    return history

# Momentum
def momentum(learning_rate, momentum, iterations):
    x = 10 # Initial guess
    v = 0
    history = [x]
    for i in range(iterations):
        v = momentum * v - learning_rate * gradient(x)
        x += v
        history.append(x)
    return history

# Adam (Adaptive Moment Estimation)
def adam(learning_rate, beta1, beta2, epsilon, iterations):
    x = 10 # Initial guess
    m = 0
    v = 0
    history = [x]
    for i in range(iterations):
        grad = gradient(x)
        m = beta1 * m + (1 - beta1) * grad
        v = beta2 * v + (1 - beta2) * (grad ** 2)
        m_hat = m / (1 - beta1**(i+1))
        v_hat = v / (1 - beta2**(i+1))
        x -= learning_rate * m_hat / (np.sqrt(v_hat) + epsilon)
        history.append(x)
    return history

# RMSprop (Root Mean Square Propagation)
def rmsprop(learning_rate, decay_rate, epsilon, iterations):
    x = 10 # Initial guess
    rmsprop_cache = 0
    history = [x]
    for i in range(iterations):
        grad = gradient(x)
        rmsprop_cache = decay_rate * rmsprop_cache + (1 - decay_rate) * grad**2
        x -= learning_rate * grad / (np.sqrt(rmsprop_cache) + epsilon)
        history.append(x)
    return history

# Define parameters
learning_rate = 0.1
iterations = 100
batch_size = 32
momentum_value = 0.9
beta1 = 0.9
beta2 = 0.999
epsilon = 1e-8
decay_rate = 0.9

# Run optimization algorithms
gd_history = gradient_descent(learning_rate, iterations)
sgd_history = stochastic_gradient_descent(learning_rate, iterations)
mbgd_history = mini_batch_gradient_descent(batch_size, learning_rate, iterations)
momentum_history = momentum(learning_rate, momentum_value, iterations)
adam_history = adam(learning_rate, beta1, beta2, epsilon, iterations)
rmsprop_history = rmsprop(learning_rate, decay_rate, epsilon, iterations)

# Plot optimization curves
plt.figure(figsize=(12, 8))
plt.plot(gd_history, label='Gradient Descent')
plt.plot(sgd_history, label='Stochastic Gradient Descent', linestyle="-.")
plt.plot(mbgd_history, label='Mini-batch Gradient Descent')
plt.plot(momentum_history, label='Momentum')
plt.plot(adam_history, label='Adam')
plt.plot(rmsprop_history, label='RMSprop')
plt.title('Optimization Methods Comparison')
plt.xlabel('Iterations')
plt.ylabel('Parameter Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data for linear regression
np.random.seed(0)
n_samples = 100
X1 = 2 * np.random.rand(n_samples, 1)  # Feature 1
X2 = 3 * np.random.rand(n_samples, 1)  # Feature 2
X_b = np.c_[np.ones((n_samples, 1)), X1, X2]  # Add a column of ones for bias term
theta_true = np.array([[3], [4], [5]])  # True parameters (including bias term)
y = X_b.dot(theta_true) + np.random.randn(n_samples, 1)  # Generate target values

# Define cost function for linear regression (mean squared error)
def compute_cost(X, y, theta):
    m = len(y)
    predictions = X.dot(theta)
    cost = (1 / (2 * m)) * np.sum(np.square(predictions - y))
    return cost

# Gradient descent function
def gradient_descent(X, y, theta, learning_rate, n_iterations):
    m = len(y)
    cost_history = np.zeros(n_iterations)

    for i in range(n_iterations):
        gradients = (1 / m) * X.T.dot(X.dot(theta) - y)
        theta = theta - learning_rate * gradients
        cost_history[i] = compute_cost(X, y, theta)

    return theta, cost_history

# Initialize parameters and hyperparameters
theta = np.random.randn(3, 1)  # Random initialization of parameters
learning_rate = 0.01
n_iterations = 1000

# Run gradient descent
theta_optimized, cost_history = gradient_descent(X_b, y, theta, learning_rate, n_iterations)

# Plot cost function over iterations
plt.plot(range(1, n_iterations + 1), cost_history, color='blue')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Gradient Descent: Cost Function Optimization')
plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Generate random data for two features
np.random.seed(0)
num_samples = 100
X1 = np.random.rand(num_samples)
X2 = np.random.rand(num_samples)
Y = 2 * X1 + 3 * X2 + np.random.randn(num_samples)  # Linear relationship with noise

# Define the cost function for linear regression
def cost_function(X1, X2, Y, theta0, theta1, theta2):
    predictions = theta0 + theta1 * X1 + theta2 * X2
    squared_errors = (predictions - Y) ** 2
    return np.mean(squared_errors) / 2

# Define the gradient descent function
def gradient_descent(X1, X2, Y, learning_rate, num_iterations):
    # Initialize theta values
    theta0 = 0
    theta1 = 0
    theta2 = 0
    
    # Lists to store optimization track
    theta0_track = [theta0]
    theta1_track = [theta1]
    theta2_track = [theta2]
    cost_track = [cost_function(X1, X2, Y, theta0, theta1, theta2)]
    
    # Gradient descent loop
    for _ in range(num_iterations):
        # Compute gradients
        predictions = theta0 + theta1 * X1 + theta2 * X2
        errors = predictions - Y
        gradient0 = np.mean(errors)
        gradient1 = np.mean(errors * X1)
        gradient2 = np.mean(errors * X2)
        
        # Update parameters
        theta0 -= learning_rate * gradient0
        theta1 -= learning_rate * gradient1
        theta2 -= learning_rate * gradient2
        
        # Append new parameters and cost to track lists
        theta0_track.append(theta0)
        theta1_track.append(theta1)
        theta2_track.append(theta2)
        cost_track.append(cost_function(X1, X2, Y, theta0, theta1, theta2))
    
    return theta0_track, theta1_track, theta2_track, cost_track

# Perform gradient descent
learning_rate = 0.01
num_iterations = 100
theta0_track, theta1_track, theta2_track, cost_track = gradient_descent(X1, X2, Y, learning_rate, num_iterations)

# Create a meshgrid of theta values
theta0_vals = np.linspace(-5, 5, 100)
theta1_vals = np.linspace(-5, 5, 100)
theta2_vals = np.linspace(-5, 5, 100)
theta0_mesh, theta1_mesh, theta2_mesh = np.meshgrid(theta0_vals, theta1_vals, theta2_vals, indexing='ij')

# Compute cost values for each combination of theta values
cost_vals = np.zeros_like(theta0_mesh)
for i in range(len(theta0_vals)):
    for j in range(len(theta1_vals)):
        for k in range(len(theta2_vals)):
            cost_vals[i, j, k] = cost_function(X1, X2, Y, theta0_vals[i], theta1_vals[j], theta2_vals[k])

# Flatten the meshgrids and cost values
theta0_vals_flat = theta0_mesh.flatten()
theta1_vals_flat = theta1_mesh.flatten()
theta2_vals_flat = theta2_mesh.flatten()
cost_vals_flat = cost_vals.flatten()

# Create the optimization track line
optimization_track = go.Scatter3d(x=theta0_track, y=theta1_track, z=cost_track, mode='lines+markers', 
                                  line=dict(color='red', width=2), marker=dict(color='red', size=5), name='Optimization Track')

# Create the surface plot of the cost function
cost_surface = go.Surface(x=theta0_vals, y=theta1_vals, z=cost_vals[:,:,0], colorscale='Viridis', opacity=0.8, name='Cost Function')

# Define layout for the plot
layout = go.Layout(title='Cost Function with Optimization Track', scene=dict(xaxis=dict(title='Theta0'), 
                                                                             yaxis=dict(title='Theta1'), 
                                                                             zaxis=dict(title='Cost')))

# Create the figure
fig = go.Figure(data=[cost_surface, optimization_track], layout=layout)

# Show the plot
fig.show()
```

**Linear Regression model prediction**

$\hat{y}= \theta_0+\theta_1 x_1+ \theta_2 x_2 + ...+\theta_n x_n$

- $\hat{y}$ is the predicted value
- $n$ is the number of features
- $x_i$ is the $i^{\text{th}}$ feature value
- $\theta_j$ is the jth model parameter, including the bias term $\theta_0$ and the feature weights $\theta_1, \theta_2, ....\theta_n$.

**Linear Regression model prediction (vectorized form)**

$\hat{y}= h_{\boldsymbol{\theta}}(\textbf{x}) = \boldsymbol{\theta} \cdot \textbf{x}$

- $h_\theta$ is the hypothesis function, using the model parameters $\theta$.
- $\boldsymbol{\theta}$ is the model's parameters vector, containing the bias term $\theta_0$ and the feature weights $\theta_1$ to $\theta_n$.
- $\textbf{x}$ is the instances's feature vector.

**MSE cost function for a linear regression function:**

$\text{MSE} = \frac{1}{m}\sum_{i=1}^m (\hat y^{(i)} - y^{(i)})^2$

or 

$\text{MSE}(\textbf{X}, h_{\boldsymbol{\theta}}) = \frac{1}{m}\sum_{i=1}^m ({\boldsymbol{\theta}}^T x^{(i)} - y^{(i)})^2$

**Normal equation:**

$\hat{\theta}= (X^T X)^{-1} X^T y$

where:
- $\hat{\theta}$ is the value of $\boldsymbol{\theta}$ that minimizes the cost function.
- $\textbf{y}$ is the vector of target values containing $y^{(1)}$ to $y^{(m)}$.




```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
m = 100  # number of instances
X = 2 * np.random.rand(m, 1)  # column vector
y = 4 + 3 * X + np.random.rand(m, 1)  # column vector

plt.scatter(X, y)  # Use scatter() function to create a scatter plot
plt.xlabel("X")
plt.ylabel("y")
plt.show()  # Display the plot
```

```python
# Now let's calculate hat_theta using the normal equation
from sklearn.preprocessing import add_dummy_feature 
X_b = add_dummy_feature(X)         # adding x_0 =1 in the matrix
theta_best = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
```

```python
X_new_b = add_dummy_feature(X_new)   # adding x_0 =1 in the matrix
```

```python
y_preict = X_new_b @ theta_best
```
